# Data preprocessing, training

```python
X = df.drop('price', axis=1)
y = df['price'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

### Scaling

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.transform(X_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
```

```python
X_train.shape
```

We have 19 incoming features, so we can have 19 neurons in our layer.

```python
model = Sequential()

model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))
model.add(Dense(19, activation='relu'))

model.add(Dense(1))

model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(x=X_train, y=y_train,
          validation_data=(X_test, y_test),
          batch_size=128,
          epochs=400)
```

# Validate and evaluate the model

See the history of the model's losses:

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses.plot()
```

If you would see a large spike in the plot above after the curve starts to flat out, it means that the model overfitted and the loss shows up on the test data.

In that case you would want to implement early stopping.

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, explained_variance_score
```

```python
predictions = model.predict(X_test)
```

```python
mean_squared_error(y_test, predictions)
```

```python
np.sqrt(mean_squared_error(y_test, predictions)) # the root mean squared error
```

```python
mean_absolute_error(y_test, predictions)
```

Check if the error is good or bad relative to the dataset:

```python
df['price'].describe()
```

```python
5.402966e+05 # the mean label value
```

```python
explained_variance_score(y_test, predictions)
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(y_test, predictions)
plt.plot(y_test, y_test, 'r')
```

In a "perfect case" the plot above would be a straight line.

But as the explained variance score also tries to tell that the model performs relatively well on the bottom 99% of houses, it makes big errors on the very expensive houses only. It would worth to retrain the model on that bottom 99% of data (houses below $3M).
