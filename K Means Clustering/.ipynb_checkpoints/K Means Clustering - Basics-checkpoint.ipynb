```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

K Means Clustering is another unsupervised learning algorithm. The idea behind it is to discover clusters of similar datapoints in the dataset.

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, centers=4, cluster_std=1.8, random_state=101)
```

```python
data
```

```python
data[0].shape
```

```python
plt.scatter(data[0][:, 0], data[0][:, 1])
```

```python
#the cluster where the datapoints belong to
data[1]
```

```python
plt.scatter(data[0][:, 0], data[0][:, 1], c=data[1])
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
#the cluster center points the model believes to be true
kmeans.cluster_centers_
```

```python
#the labels (cluster number) the model believes to be true (predicted)
kmeans.labels_
```

```python
#the actual (correct) labels for the generated data
data[1]
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, sharey=True, figsize=(10, 6)) #tuple unpacking, btw

ax1.set_title('K Means')
ax1.scatter(data[0][:, 0], data[0][:, 1], c=kmeans.labels_)

ax2.set_title('Original')
ax2.scatter(data[0][:, 0], data[0][:, 1], c=data[1])
```
