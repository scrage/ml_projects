```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
iris = sns.load_dataset('iris')
iris.head()
```

```python
iris['species'].unique()
```

```python
sns.pairplot(iris)
```

# Pair Grid

PairGrid gives an empty pair grid for each column in the dataset. It takes all the numerical columns and grids them up. `pairplot` is a simplified version of `PairGrid`, the latter giving a lot more control.

The main way of using Seaborn's PairGrid is to set it to a variable, then off of that variable we can map plot types.

```python
g = sns.PairGrid(iris)
g.map_diag(sns.histplot)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)
```

# Facet Grid

`FacetGrid` is the general way to create grids of plots based off of a feature.

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
```

Again, we can map plot types to this -by default- blank grid.

The row and column names are what variables we want to separate by for the facet grid, and the 3rd argument in play is the variable we actually want to map to for display.

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(sns.histplot, 'total_bill')
```

If a plot type is needed that has more than one variable, then those extra variables need to be provided to the `map` function.

```python
g = sns.FacetGrid(data=tips, col='time', row='smoker')
g.map(plt.scatter, 'total_bill', 'tip')
```
