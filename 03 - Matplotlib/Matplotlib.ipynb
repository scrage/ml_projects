```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

There are 2 ways or approaches to create plots with matplotlib.

# FUNCTIONAL way of creating matplotlib plots

```python
plt.plot(x, y)
#plt.show()
```

Adding labels an title:

```python
plt.plot(x, y)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('Title')
```

Multiplots on the same canvas:

```python
plt.subplot(1, 2, 1)
plt.plot(x, y, 'r')

plt.subplot(1, 2, 2)
plt.plot(y, x, 'b')
```

# Object-Oriented way of creating matplotlib plots

```python
fig = plt.figure()

axes = fig.add_axes([0.1, 0.2, 0.8, 0.8]) #left, bottom, width and height of the axis

axes.plot(x, y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

### 2 sets of figures on one canvas

```python
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) #the bottom of the inserted axes starts 20% in from the left, 50% up from the bottom,
#takes up 40% of the canvas' width and 30% of the canvas' height

axes1.plot(x, y)
axes1.set_title('Larger plot')

axes2.plot(y, x)
axes2.set_title('Smaller plot')
```

## Creating subplots the OO way

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

#axes.plot(x, y)
plt.tight_layout() #fixes overlaps in display

for current_ax in axes:
    current_ax.plot(x, y)
```

`axes` here is a list of exes objects, meaning we can iterate through them

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y)
axes[0].set_title('First plot')

axes[1].plot(y, x)
axes[1].set_title('Second plot')

plt.tight_layout()
```

## Figure size and DPI

`figsize` is a tuple with the width and height of the figure _in inches_.


```python
fig = plt.figure(figsize=(3, 2), dpi=100)

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y)

ax.set_title('Title')
ax.set_xlabel('X label')
ax.set_ylabel('Y label')
```

Same applies with subplots:

```python
fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 2), dpi=100)

axes[0].plot(x, y)
axes[1].plot(y, x)

plt.tight_layout()
```

## Legends

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2)
ax.plot(x, x**3)

ax.legend()
#if the label parameter is not provided for the axes, then a runtime warning will come:
```

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='X squared')
ax.plot(x, x**3, label='X cubed')

ax.legend()
```

We can define the location of the legend as well.
See the [documentation page](http://matplotlib.org/users/legend_guide.html#legend-location) for details.

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, x**2, label='X squared')
ax.plot(x, x**3, label='X cubed')

ax.legend(loc=10)
```

Some of the most common **loc** values are:

```python
ax.legend(loc=1) # upper right corner
ax.legend(loc=2) # upper left corner
ax.legend(loc=3) # lower left corner
ax.legend(loc=4) # lower right corner

# .. many more options are available

# Most common to choose
ax.legend(loc=0) # let matplotlib decide the optimal location
fig
```

## Setting colours

```python
fig = plt.figure()

ax = fig.add_axes([0, 0, 1, 1])

ax.plot(x, y, color='orange')
```

RGB hashcodes can be used as well.

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='#ebedf5')
```

## Line width and line style

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='green', linewidth=4, alpha=0.25) #linewidth=1 is the default width - lw is short for this parameter:
ax.plot(x, y, color='green', lw=5, alpha=0.25)
```

`linestyle='--'`:

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=3, alpha=0.5, linestyle='--')
```

`ls` is short for `linestyle`

`linestyle='-.'`:

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=3, alpha=0.5, ls='-.')
```

`ls=':'`:

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=3, alpha=0.5, ls=':')
```

Other supported values are `'-', '--', '-.', ':', 'None', ' ', '', 'solid', 'dashed', 'dashdot', 'dotted'`.

## Markers

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=1, alpha=0.5, ls='-', marker='o')
```

Available marker codes: https://matplotlib.org/api/markers_api.html?highlight=marker

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='orange', lw=1, alpha = 0.7, ls='-', marker='p', markersize=15, markerfacecolor='red', markeredgewidth=3, markeredgecolor='blue')
```

## Plot range

We can show the plot between a certain range on either of the axes:

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=2, ls='--')

ax.set_xlim([0, 1])
```

```python
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.plot(x, y, color='purple', lw=2, ls='--')

ax.set_xlim([0, 1])
ax.set_ylim([0, 2])
```
